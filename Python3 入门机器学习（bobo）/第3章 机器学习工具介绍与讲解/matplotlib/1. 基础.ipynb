# matplotlib 基础

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy
```

```python
plt.plot(x, y)
plt.plot(x, cosy, color='r', linestyle='-.')
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, cosy, color='r', linestyle='-.')
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, cosy, color='r', linestyle='-.')
# plt.xlim(0, 10)
# plt.ylim(-1, 1)
plt.axis((-1, 11, -2, 2))
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, cosy, color='r', linestyle='-.')
# plt.xlim(0, 10)
# plt.ylim(-1, 1)
plt.axis((-1, 11, -2, 2))
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, y, label="sin(x)")
plt.plot(x, cosy, color='r', linestyle='-.', label="cos(x)")
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.legend()
plt.show()
```

```python
%%time
plt.plot(x, y, label="sin(x)")
plt.plot(x, cosy, color='r', linestyle='-.', label="cos(x)")
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.legend()
plt.title("Welcome to the machine learning")
plt.show()
```

## Scatter Plot
