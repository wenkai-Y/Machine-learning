# 2.PCA过程：
## 第一步：将样本的均值归0（demean）
+  $ Var(x) = \frac{1}{m}\sum_{i=1}^m (x_i)^2 $
+ $ \overline x = 0$

## 第二步：我们想要求一个轴的方向
+  $w = (w_1, w_2)$

## 第三步：使得我们所有的样本，映射到$w$之后，有：
- $ MAX: Var(X_{project}) = \frac{1}{m}\sum_{i=1}^m (X_{project}^{(i)} - \overline X_{project})^2 $
- $MAX: Var(X_{project}) = \frac{1}{m}\sum_{i=1}^m \mid \mid X_{project}^{(i)} - \overline X_{project}\mid \mid^2 $
- $MAX: Var(X_{project}) = \frac{1}{m}\sum_{i=1}^m \mid \mid X_{project}^{(i)}\mid \mid^2$
